```python
#EJECUTAR EN DEPNOTE
import os
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import TimeSeriesSplit
import warnings
warnings.filterwarnings('ignore', category=pd.core.common.SettingWithCopyWarning)
warnings.filterwarnings('ignore', category=UserWarning)
os.chdir(r'/work/DataSets_Codigo')
df_filtrado = pd.read_csv('Mer_Filtrado_sin_2024.txt', encoding="latin1")
df_filtrado['Fec'] = pd.to_datetime(df_filtrado['Fec'])
df_filtrado['Mes'] = df_filtrado['Fec'].dt.month
df_filtrado['AÃ±o'] = df_filtrado['Fec'].dt.year
df_filtrado['Trimestre'] = df_filtrado['Fec'].dt.quarter
df_filtrado['DiaSemana'] = df_filtrado['Fec'].dt.weekday
df_filtrado['FinDeSemana'] = (df_filtrado['DiaSemana'] >= 5).astype(int)
ventas_mensuales = df_filtrado.groupby(['AÃ±o', 'Mes'])['Cant'].sum().reset_index()
ventas_mensuales_agregadas = df_filtrado.groupby(['Mes']).agg({'Cant': 'sum'}).reset_index()
ventas_mensuales_agregadas.rename(columns={'Cant': 'Ventas_Hist'}, inplace=True)
ventas_mensuales['Ventas_Movil'] = ventas_mensuales['Cant'].rolling(window=3, min_periods=1).mean()
Q1 = ventas_mensuales.quantile(0.25)
Q3 = ventas_mensuales.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (ventas_mensuales < lower_bound) | (ventas_mensuales > upper_bound)
ventas_mensuales_filt = ventas_mensuales[~outliers.any(axis=1)]
tscv = TimeSeriesSplit(n_splits=5)
best_params_xgb = {'n_estimators': 300, 'max_depth': 10}
def add_product_sales_history_features(df, history_months=[1, 3, 6]):
    for month in history_months:
        df[f'Sales_Last_{month}_Months'] = df['Cant'].shift().rolling(window=month).sum().fillna(0)
    return df
ventas_mensuales_filt = add_product_sales_history_features(ventas_mensuales_filt)
predicciones_2024_xgb = []
for mes in range(1, 13):
    df_mes = ventas_mensuales_filt[ventas_mensuales_filt['Mes'] == mes]
    if not df_mes.empty:
        X = df_mes[['AÃ±o', 'Ventas_Movil', 'Sales_Last_1_Months', 'Sales_Last_3_Months', 'Sales_Last_6_Months']].copy()  
        y = df_mes['Cant']
        xgb_model = xgb.XGBRegressor(**best_params_xgb, random_state=42)
        xgb_model.fit(X, y)
        ventas_movil_mes = df_mes['Ventas_Movil'].values[0]
        ventas_pred_xgb = xgb_model.predict([[2024, ventas_movil_mes] + list(df_mes.iloc[-1, -3:])])
        ventas_pred_enteras_xgb = int(round(ventas_pred_xgb[0]))
        predicciones_2024_xgb.append({'AÃ±o': 2024, 'Mes': mes, 'Ventas': ventas_pred_enteras_xgb})
    else:
        print(f"No hay datos disponibles para el mes {mes}")
predicciones_2024_xgb = pd.DataFrame(predicciones_2024_xgb)
print(predicciones_2024_xgb)
```
